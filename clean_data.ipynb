## Transformation of NaN values

**First option**: parsing of the numerical column 'salary' using the mean.

```python
df_without_null['salary']
```

```python
# Extract de numerical float value of the salary

df_without_null['salaries_numerical'] = df_without_null['salary'].replace('[^\d.]', '', regex=True).str.extract('(\d+)').astype(float)
```
